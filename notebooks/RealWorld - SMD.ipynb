## REAL WORLD DATA -> SMD

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

def load_smd(name):
    data = np.loadtxt(f"../data/smd/test/{name}.txt", delimiter=",")
    labels = np.loadtxt(f"../data/smd/test_label/{name}.txt", delimiter=",")
    return data, labels

x, y = load_smd("machine-1-1")
```

```python
def plot_data(x, y):    
    size = 1 if len(x.shape) == 1 else x.shape[1]
    fig = plt.Figure(figsize=(15, 4*size))    
    labels_idx = np.where(y == 1)[0]  
    for i in range(size):
        ax = fig.add_subplot(size, 1, i+1)
        for label in labels_idx:
            ax.axvline(label, color='aquamarine', alpha=0.3)
        ax.plot(x[:, i])
    return fig
```

```python
plot_data(x[:, :2], y)
```
